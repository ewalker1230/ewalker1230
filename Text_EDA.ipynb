```python
words = nltk.word_tokenize(text_desc)
```

```python
words
```

```python
# set english stopwords
stopwords_eng = set(nltk.corpus.stopwords.words('english'))

# Remove single-character tokens (mostly punctuation)
words = [word for word in words if len(word) > 1]


# Remove numbers
words = [word for word in words if not word.isnumeric()]

# Lowercase all words (default_stopwords are lowercase too)
words = [word.lower() for word in words]

# remove stopwords
words = [word for word in words if word not in stopwords_eng]

```

```python
word_pos = nltk.pos_tag(words)
word_pos_15 = word_pos[0:16]
word_pos_15
```

```python
word_pos_df = pd.DataFrame(word_pos)
word_pos_df = word_pos_df.rename(columns={0:"words",1:"part_of_speech"})
word_pos_df

```
